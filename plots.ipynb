```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,4*np.pi,100)
# print(x)

sin_y = np.sin(x)
cos_y=np.cos(x)


plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cosine')

plt.legend(loc='upper right')
plt.title("Sine Wave")
plt.title("Cos wave")
plt.show()
```

```python
x = np.linspace(0,4*np.pi,100)
# print(x)

sin_y = np.sin(x)
cos_y=np.cos(x)


plt.plot(x,sin_y,'-')
plt.plot(x,cos_y,'*')

# plt.legend(loc='upper right')
plt.title("Sine Wave")
plt.title("Cos wave")
plt.show()
```

```python
plt.style.use('tableau-colorblind10')
plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cosine')

# plt.legend(loc='upper right')
plt.title("Sine Wave")
plt.title("Cos wave")

plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(28,4))
plt.style.use('tableau-colorblind10')
plt.plot(x,sin_y,label='Sine',color='red',linewidth=5)
plt.plot(x,cos_y,'*',label='Cosine',color='white')

# plt.legend(loc='upper right')
plt.title("Sine Wave")
plt.title("Cos wave")
plt.legend()


fig.savefig('sin.png')
# plt.show()
```

```python
 from sklearn.datasets import fetch_california_housing

dataset = fetch_california_housing()
dataset.keys()
target = dataset['target'][:10]
print(target)
```

```python
plt.figure(figsize=(18,5))
plt.style.use('tableau-colorblind10')
plt.xlim([0,2])
plt.plot(target,marker='o',linestyle='dashed',color='red')
plt.title('California Housing Data')
plt.show()
```

```python
x1= np.random.randn(10000)
y1=np.random.randn(10000)

x2=x1+2
y2=y1+2
plt.figure(figsize=(12,5))
plt.style.use('seaborn-v0_8-dark-palette')
plt.scatter(x1,y1,c='red',marker='o',alpha=0.5,label="+0")
plt.scatter(x2,y2,c='blue',marker='o',alpha=0.5,label='+1')
plt.legend(loc='lower left')
plt.show()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.feature_names)
```

```python
plt.style.use('tableau-colorblind10')

plt.figure(figsize=(10,4))
x = iris['data'][:,0]
y = iris['data'][:,1]
plt.scatter(x,y,s=100,marker='o',alpha=0.5,c=x)
plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
plt.colorbar()
plt.show()
```
