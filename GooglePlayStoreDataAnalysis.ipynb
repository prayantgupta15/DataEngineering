**Total number apps for each category.**

```python
categories = df['Category'].unique()

dc = {}
for c in categories:
  fd = df[df['Category']==c]
  l = len(fd)
  dc[c]=l

dc
```

**Total number of apps in each Type**


```python
types = df['Type'].unique()
mp ={}

for t in types:
  apps = df[df['Type']==t]
  mp[t]= len(apps)

print(mp)

df['Type'].describe()

df.groupby('Type').groups
```
